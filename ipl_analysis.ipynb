```python
query="""SELECT winner, COUNT(*) AS successful_chases
FROM matches
WHERE win_by_wickets > 0
GROUP BY winner
ORDER BY successful_chases DESC
LIMIT 1;"""
print(pd.read_sql(query,conn))
```

```python
query = """
SELECT m.season, m.winner, 
       COUNT(*) AS matches_won,
       ROUND((COUNT(*) * 100.0) / 
             (SELECT COUNT(*) FROM matches 
              WHERE season = m.season AND (team1 = m.winner OR team2 = m.winner)), 2) 
       AS win_percentage
FROM matches m
WHERE m.winner IS NOT NULL
GROUP BY m.season, m.winner
ORDER BY m.season, win_percentage DESC;
"""
print(pd.read_sql(query,conn))
```

```python
import pandas as pd
matches = pd.read_csv(r"C:\Users\HP\Downloads\matches.csv")
deliveries=pd.read_csv(r"C:\Users\HP\Downloads\deliveries.csv.zip")
won = matches.groupby(["season", "winner"]).size().reset_index(name="Matches_won")
print(won)
sns.lineplot(data=won, x="season", y = "Matches_won", hue="winner", marker = 'o')
plt.xlabel("Season", fontsize=12)
plt.ylabel("Matches_Won", fontsize=12)
plt.title("Number of Matches Won by Each Team Over Time", fontsize=16)
plt.legend(title="Team")
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
if "batsman_runs" not in deliveries.columns or "ball" not in deliveries.columns:
    print("Error: Required columns not found in dataset")
else:
    total_runs = deliveries.groupby("batter")["batsman_runs"].sum()
    balls_faced = deliveries.groupby("batter")["ball"].count()
    strike_rate = (total_runs / balls_faced.replace(0, np.nan)) * 100
    valid_dismissals = deliveries[deliveries["player_dismissed"].notna()].groupby("player_dismissed").size()
    batting_avg = total_runs / valid_dismissals.reindex(total_runs.index).fillna(0)
    batting_avg.replace([float('inf'), 0], float('nan'), inplace=True)  # Handle infinity

    batting_stats = pd.DataFrame({"Batting Average": batting_avg, "Strike Rate": strike_rate}).dropna()
    top_batters = batting_stats.nlargest(20, "Batting Average")
    plt.figure(figsize=(10, 6))
    sns.scatterplot(
        x=top_batters["Strike Rate"], 
        y=top_batters["Batting Average"], 
        hue=top_batters.index, 
        s=150, edgecolor="black"
    )
    plt.xlabel("Strike Rate", fontsize=12)
    plt.ylabel("Batting Average", fontsize=12)
    plt.title("Strike Rate vs. Batting Average of Top Batters", fontsize=14)
    plt.legend(title="Batter", bbox_to_anchor=(1, 1), fontsize=10)
    plt.grid(True)
    plt.show()
```

```python
stadium_wins = matches.groupby("venue")["winner"].value_counts().unstack().fillna(0)
top_stadium = stadium_wins.sum(axis=1).idxmax()

print(f"The stadium with the most games won: {top_stadium}")
print(stadium_wins.loc[top_stadium].sort_values(ascending=False))
```

```python

runs_conceded = deliveries.groupby("bowler")["total_runs"].sum()
overs_bowled = deliveries.groupby("bowler")["ball"].count() / 6
economy_rate = (runs_conceded / overs_bowled).dropna()
top_10_bowlers = economy_rate.nsmallest(10)
plt.figure(figsize=(10, 5))
sns.barplot(x=top_10_bowlers.index, y=top_10_bowlers.values, palette="viridis")
plt.xlabel("Bowler")
plt.ylabel("Economy Rate")
plt.title("Top 10 Most Economical Bowlers")
plt.xticks(rotation=45)
plt.show()
```
